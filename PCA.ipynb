```python
!pip install scikit-learn
import numpy as np
import pandas as pd
import os
import sklearn
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import gc 
```

```python
# Define simple function to run PCA
def PCA_run(x, n_components):
    pca = PCA(n_components=n_components)
    x_pca = pca.fit_transform(x)
    tot_var = sum(pca.explained_variance_ratio_)
    return x_pca, tot_var


'''
Function loops through a list of component numbers, runs PCA with each number of components
and saves each to a csv file.

For example, inputting the list [70,80,90] would results in three PCA's being executed with 
70,80,and 90 principal components, respectively. The resulting vectors would be saved to three separate
csv files, titled pc70.csv, pc80.csv and pc90.csv, respectively.

Inputs: list of number of components, list of vectors on which to perform PCA, list of SMILES strings
corresponding to the inputted vectors

Returns: List containing the total variance explained with the inputted number of principal components

'''
def pcaLoop(numPC, vectors, allSmilesFiltered):
    totVarList = []
    for num in numPC:
        print(num)
        numString = str(num)
        pcaComplete = PCA_run(vectors, num)
        totVarList.append(pcaComplete[1])
        
        
        print("variance")
        print(pcaComplete[1])

        detectedPCA = pcaComplete[0]
        string_indices = []
        for i in range(len(detectedPCA)):
            knn_string = ', '.join(str(k) for k in detectedPCA[i])
            string_indices.append(knn_string)

        bracket_string_indices = []
        for i in range(len(string_indices)):
            bracket_string = '[' + string_indices[i] + ']'
            bracket_string_indices.append(bracket_string)
        
        
        headerString = "pc" + numString
        
        saveString = headerString + ".csv"
        
        pcaDF = pd.DataFrame()
        pcaDF['smiles'] = allSmilesFiltered
        pcaDF[headerString] = bracket_string_indices
        savePath = os.path.join(os.getcwd(), saveString)
        pcaDF.to_csv(savePath)
        
        del pcaDF
        del string_indices
        del bracket_string_indices
        del pcaComplete
        
        gc.collect()
        
    return totVarList
```
